```python
import pandas as pd
```

```python
data = pd.DataFrame({
    'Name':        ['Harry', 'Draco', 'Ron',   'Tom',   'Albus'],
    'Allegiance':  ['Phoenix', 'Death Eater', 'Phoenix', 'Death Eater', 'Phoenix'],
    'Approx Age':  ['17',      '17',          '17',      '55',           '103']})

data
```

```python
data.info()
```

```python
data['Approx Age'] = data['Approx Age'].astype(int)

data.info()
```

```python
data['age_minus5'] = data['Approx Age'].apply(lambda x: x - 5)

data
```

```python
def flag_group(row):
    if row['Allegiance'] == 'Phoenix':
        if row['Approx Age'] > 18:
            return 'Order of Phoenix'
        else:
            return "Dumbledore's Army"
    return 'Death Eater'

data['profile_flag'] = data.apply(flag_group, axis=1)

data
```

```python
data['profile_flag_2'] = data.apply(
     lambda r: 'Order of Phoenix' if (r['Approx Age'] > 18 and r['Allegiance']=='Phoenix')
               else ("Dumbledore's Army" if (r['Approx Age'] <= 18 and r['Allegiance']=='Phoenix')
                     else 'Death Eater'),
     axis=1
 )
data
```
